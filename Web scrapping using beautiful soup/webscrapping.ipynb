```python
from urllib.request import urlopen
android_url="https://en.wikipedia.org/wiki/Android_version_history"
```

```python
android_data=urlopen(android_url)
print(type(android_data))
```

```python
android_html=android_data.read()
```

```python
from bs4 import BeautifulSoup as soup
android_soup=soup(android_html,'html.parser')
```

```python
type(android_soup)
```

```python
android_soup.h1
```

```python
table=android_soup.findAll('table',{'class':'wikitable'})
len(table)
```

```python
android_table=table[0]
```

```python
header= android_table.findAll('th')
len(header)
```

```python
coloumn_titles=[ct.text[0:-1] for ct in header]
coloumn_titles
```

```python
rows_data=android_table.findAll('tr')[1:]
first_row=rows_data[0].findAll('td',{})
print(len(first_row))
for i in first_row:
    print(i.text[:-1])
```

```python
final_rows=[]
for  i in rows_data:
    current_row=[]
    f=i.findAll('td')
    if len(f)==7:
        name=f[0]
        int_name=f[1]
    elif len(f)<7:
        f.insert(0,name)
        f.insert(1,int_name)        
    for ind,j in enumerate(f):
        if(ind==3):
            date=j.text[:-1].split(',')
            date_str='-'.join(date)
            current_row.append(date_str)
        else:
            current_row.append(j.text[:-1])
        
    final_rows.append(current_row)
    
    
```

```python
##creating csv file 
filename= 'android_version_history.csv'
with open(filename,'w',encoding='utf-8')as f:
    header_string=','.join(coloumn_titles)
    header_string+='\n'
    f.write(header_string)
    for i in final_rows:
        current_row_string=','.join(i)
        current_row_string+='\n'
        f.write(current_row_string)
            
```
